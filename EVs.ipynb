## Exploratory Data Analysis on EVs

### Importing Packages

```python
import numpy as np 
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Importing the data

```python
evdata= pd.read_csv('ElectricCarData_Clean.csv')
evdata
```

### Checking if null value exists

```python
evdata.isnull().sum()
```

### Describing the dataset

```python
evdata.describe()
```

```python
evdata.info()
```

```python
a=np.arange(1,104)
```

### Frequency of the brands

```python
plots= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y=a,data=evdata, palette='viridis')
plt.grid(axis='y')
plt.title('Brands in the datset')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=53)
```

### Top speed achieved by a car brand

```python
speed_plots = plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y='TopSpeed_KmH',data=evdata,palette='magma')
plt.grid(axis='y')
plt.title('Top Speed achieved by a car brand')
plt.xlabel('Brand')
plt.ylabel('Top Speed')
plt.xticks(rotation=53)
```

### Price of Cars (in Euros)

```python
price_plots= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y='PriceEuro',data=evdata,palette='Blues')
plt.title('Price of a Car')
plt.xlabel('Price in Euro')
plt.grid(axis='y')
plt.ylabel('Frequency')
plt.xticks(rotation=53)
```

### Number of seats

```python
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightseagreen']

plt.figure(figsize=(7,7))
plt.pie(
    evdata['Seats'].value_counts(),
    labels=evdata['Seats'].value_counts().index,
    colors=colors, autopct='%0.1f%%', pctdistance=0.85, 
    explode=(0.05, 0.05, 0.05, 0.05, 0.05), 
    startangle=90 )

plt.gca().add_artist(plt.Circle((0,0), 0.6, fc='white'))

plt.axis('equal')
plt.title('Seats')
plt.tight_layout()
plt.show()
```
